```python
# Import Libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create
```

```python
# Read in the data

df = pd.read_csv(r"C:\Users\SAMUEL\Documents\Alex The Analyst\Data\movies.csv")
df.head()
```

```python
# Checking for missing data in each columns
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# data types for each columns
df.dtypes
```

```python
# Creating a correct year column
df['YearCorrect'] = df['released'].astype(str).str[-20:]
df
```

```python
# Ordering by the gross value
# pd.set_option('display.max_rows', None) used to display all the rows in the data instead of the 5 first, 5 last
df.sort_values(by=['gross'], inplace=False, ascending=False) # inplace=True doesn't output the result
```

```python
# drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatterplot with budget vs gross revenue
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget Vs Gross')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# regression plot of budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df,
            scatter_kws={"color": "black"}, line_kws={"color": "blue"})
```

```python
# Looking at correlations
df.corr(method='kendall') #3 types of correlation: kendall, pearson, spearman
df.corr() # by default->Pearson's
```

```python
# Visualizing the correlation matrix 
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True) # annot=False makes the correlation values not to show
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# company can also be a correlation factor, but we have to numerize it
df_numerized = df

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes # assigns random unique category values to them
        
df_numerized        
```

```python
correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix, annot=True) # annot=False makes the correlation values not to show
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
